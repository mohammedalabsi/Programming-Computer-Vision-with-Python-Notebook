```python
from PIL import Image
from pylab import *
import matplotlib.pyplot as pl
# read image to array
im = array(Image.open('dsu.jpg'))
pl.rcParams["figure.figsize"] = (16,9)
# plot the image
imshow(im)
# some points
x = [120,120,2100,2100]
y = [150,2000,150,2000]
# plot the points with red star-markers
plot(x,y,'r*')
# line plot connecting the first two points
plot(x[:2],y[:2])
# add title and show the plot
title('Plotting: "dongseo.jpg"')
show()
#plot(x,y) # default blue solid line
#plot(x,y,r*) # red star-markers
#plot(x,y,go-) # green line with circle-markers
#plot(x,y,ks:) # black dotted line with square-markers
```

```python
# read image to array
im = array(Image.open('dsu.jpg').convert('L'))
# create a new figure
figure()
# dont use colors
gray()
imshow(im)
figure()
# show contours with origin upper left corner
contour(im, origin='image')
axis('equal')
axis('off')
show()
```
