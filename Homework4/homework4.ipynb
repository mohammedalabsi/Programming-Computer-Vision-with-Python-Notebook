```python
#Transforming Image 1
import cv2
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
img_1 = cv2.imread('dsu1.jpg')
pts1 = np.array([[324,280],[762,313],[325,535],[766,508]])
img_2 = cv2.imread('dsu1.jpg')
pts2 = np.array([[324,280],[762,280],[325,508],[760,508]])
h,status = cv2.findHomography(pts1,pts2)
new_img = cv2.warpPerspective(img_1,h,(img_2.shape[1],img_2.shape[0]))
fig, axs = subplots(1, 2, figsize=(16, 16))
axs[0].imshow(img_1)
axs[0].axis('off')
title('Transformed Image')
axs[1].imshow(new_img)
axs[1].axis('off')
plt.show()
```
